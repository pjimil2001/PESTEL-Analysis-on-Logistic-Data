```python
import numpy as np
import pandas as pd
pd.options.display.max_columns = None
pd.options.display.max_rows=None

# Load the dataset
file_path = "case study dataset.csv"
df = pd.read_csv(file_path)

# Display the first few rows to understand the dataset
df.head()
```

```python
df.info()
```

```python
# 1. Remove duplicate rows
df_cleaned = df.drop_duplicates()

# 2. Handling missing values: Identifying null and non-null values
null_counts = df_cleaned.isnull().sum()
not_null_counts = df_cleaned.notnull().sum()

# 3. Adding variable types
variable_types = df_cleaned.dtypes

# 4. Finding the total number of rows
total_rows = df_cleaned.shape[0]

# 5. Descriptive statistics
descriptive_stats = df_cleaned.describe()

# Displaying results
#Instead of using ace_tools, we display dataframes using Pandas functionalities

print("Null Values Count:")
display(pd.DataFrame(null_counts, columns=['Null Count']))
print("\nNon-null Values Count:")
display(pd.DataFrame(not_null_counts, columns=['Not Null Count']))
print("\nVariable Types:")
display(pd.DataFrame(variable_types, columns=['Data Type']))
print("\nDescriptive Statistics:")
display(descriptive_stats)

print("\nTotal Rows:", total_rows)
```

```python
int_columns = ['category_id',"customer_id","customer_zipcode","order_customer_id","order_item_id", 'product_card_id','order_id']

for col in int_columns:
    if col in df.columns:
        df[col] = df[col].fillna(0).astype(int)

# Quick display of updated DataFrame
print("Updated DataFrame:")
display(df.head())
```

```python
# Select float columns excluding 'longitude' and 'latitude'
float_columns = df.select_dtypes(include=['float']).columns.difference(['longitude', 'latitude'])

# Round these columns to two decimal places
df[float_columns] = df[float_columns].round(2)

# Display updated DataFrame
print("Updated DataFrame with Floats Rounded to Two Decimal Places (Longitude and Latitude unchanged):")
display(df.head())
```

```python
date_columns = ['order_date', 'shipping_date']

for col in date_columns:
    if col in df.columns:
        df[col] = df[col].astype(str).str.split(' ').str[0]

# Display the updated DataFrame
print("Updated DataFrame with Dates Only:")
display(df.head())
```

```python
df.to_csv(('case_study_dataset_cleaned.csv'), index=False)
```
